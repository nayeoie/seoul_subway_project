## Q3. 그림 ② — 승차 vs 하차 비교 (4점)

**시각화 목표:** 출근 시간대(7-9시)의 승차 핫스팟과 하차 핫스팟이 다른 역임을 보여주는 그림.

**과제:**
- (a) 시간대별 또는 역별 승차·하차 인원을 비교하는 그림 1개
  - 같은 시간대에 승차 vs 하차가 다른 역에 모임을 시각적으로 강조
  - 승차·하차가 다른 색으로 구분돼야 함
  - 예시 접근: 출퇴근 핫스팟 상위 N개 역의 막대그래프 (승차/하차 stack 또는 dodge)
- (b) `fig02_*.png` 로 저장
- (c) 캡션 작성

**평가 포인트:** 승차 우세 역과 하차 우세 역이 한눈에 구분되는가?

````python
subway.head()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

morning_subway = subway[subway['hour'].isin([7, 8])]

station_line_total = morning_subway.groupby(['line', 'station', 'type'])['value'].sum().reset_index()

top_boarding_df = station_line_total[station_line_total['type'] == '승차'].nlargest(5, 'value')
top_alighting_df = station_line_total[station_line_total['type'] == '하차'].nlargest(5, 'value')

target_pairs = set(zip(top_boarding_df['line'], top_boarding_df['station'])).union(
    set(zip(top_alighting_df['line'], top_alighting_df['station']))
)

plot_data = station_line_total[station_line_total.apply(lambda row: (row['line'], row['station']) in target_pairs, axis=1)]
plot_data['display_name'] = plot_data['line'] + ' ' + plot_data['station']

plt.figure(figsize=(12, 5))

sns.barplot(
    data=plot_data, 
    x='display_name', 
    y='value', 
    hue='type', 
    palette={'승차': '#4c72b0', '하차': '#c44e52'}
)

plt.title("출근 시간대 주요 노선별 역 승차 vs 하차 총 인원수 대조", pad=20, fontsize=40)
plt.xlabel("지하철역", fontsize=30)
plt.ylabel("총 승하차 인원수", fontsize=30)
plt.xticks(rotation=45, ha='right')

plt.savefig('fig02_morning_boarding_alighting.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()
````

## Q4. 그림 ③ — 주거지 vs 업무지 분류 시각화 (4점)

**시각화 목표:** Week 2의 commute_asymmetry 결과를 시각적으로 명료하게.

**과제:**
- (a) Week 2의 `station_classification` 결과를 시각화 (그림 종류는 팀이 결정)
  - 예시: 상위 10 주거지 + 상위 10 업무지를 한 그림에 비교 막대
  - 또는 산점도 (x: morning_boarding, y: morning_alighting, 대각선이 균형선)
  - 또는 양방향 막대그래프 (좌: 업무지, 우: 주거지)
- (b) **균형선 (asymmetry = 1.0)** 또는 **대각선** 등 분류 기준선 명시
- (c) 색상이 분류(주거지/업무지)에 따라 의미를 가지면 더 좋음
- (d) `fig03_*.png` 로 저장
- (e) 캡션 작성

````python
!pip install adjustText
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text  

morning_subway = subway[subway['hour'].isin([7, 8])]
morning_subway['type'] = np.where(morning_subway['variable'].str.contains('승차'), '승차', '하차')

asymmetry_df = morning_subway.pivot_table(
    index=['line', 'station'],
    columns='type',
    values='value',
    aggfunc='mean'
).dropna().reset_index()

asymmetry_df['commute_asymmetry'] = asymmetry_df['승차'] / asymmetry_df['하차']
asymmetry_df['총인원'] = asymmetry_df['승차'] + asymmetry_df['하차']

asymmetry_df['분류'] = np.where(asymmetry_df['commute_asymmetry'] > 1.0, '주거지 (승차 우세)', '업무지 (하차 우세)')

valid_stations = asymmetry_df[asymmetry_df['총인원'] >= 3000].copy()

top_10_residential = valid_stations[valid_stations['분류'] == '주거지 (승차 우세)'].nlargest(10, 'commute_asymmetry')
top_10_business = valid_stations[valid_stations['분류'] == '업무지 (하차 우세)'].nsmallest(10, 'commute_asymmetry')
top_20_stations = pd.concat([top_10_residential, top_10_business])

plt.figure(figsize=(13, 12))

sns.scatterplot(
    data=asymmetry_df,
    x='승차', y='하차', hue='분류',
    palette={'업무지 (하차 우세)': '#c44e52', '주거지 (승차 우세)': '#4c72b0'},
    alpha=0.5, s=60
)

sns.scatterplot(
    data=top_20_stations,
    x='승차', y='하차', hue='분류',
    palette={'업무지 (하차 우세)': '#c44e52', '주거지 (승차 우세)': '#4c72b0'},
    s=200, legend=False, edgecolor='white', linewidth=2
)

max_val = max(asymmetry_df['승차'].max(), asymmetry_df['하차'].max())
plt.plot([0, max_val], [0, max_val], color='gray', linestyle='--', linewidth=1.5, label='균형선 (Asymmetry = 1.0)')

texts = []
for _, row in top_20_stations.iterrows():
    texts.append(plt.text(
        row['승차'], row['하차'], 
        f"{row['line']} {row['station']}",  
        fontsize=10, 
        fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.9, edgecolor='gray', linewidth=0.5)
    ))

adjust_text(
    texts, 
    arrowprops=dict(arrowstyle="-", color='gray', lw=1), 
    expand_points=(1.2, 1.2),  
    force_text=(0.1, 0.2)      
)


plt.title("출근 시간대(07~09시) 주거지 vs 업무지 거점역 분포", pad=20, fontsize=30, fontweight='bold')
plt.xlabel("평균 승차 인원수", fontsize=25)
plt.ylabel("평균 하차 인원수", fontsize=25)
plt.xlim(0, max_val * 1.05)
plt.ylim(0, max_val * 1.05)


plt.savefig('fig03_commute_asymmetry.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()

print("fig03_commute_asymmetry.png 저장 완료 (300 DPI)")

exact_validation = morning_subway.pivot_table(
    index=['station', 'line'], columns='type', values='value', aggfunc='mean'
).dropna()

print(exact_validation.loc[['신림', '여의도']])
````

역을 지나갈때마다 승차인원 - 하차인원 누적 합산 계산해서 지하철에 얼마나 타고 있는지 꺾은선 그래프로 보여주기



````python
df_morning = subway[(subway['line'] == '2호선') & (subway['hour'].isin([7, 8]))].copy()

df_morning['daily_value'] = df_morning['value'] / 30

df_pivot = df_morning.pivot_table(index='station', columns='type', values='daily_value', aggfunc='sum').fillna(0)

station_sequence = [
    '신도림', '대림(구로구청)', '구로디지털단지', '신대방', '신림',
    '봉천', '서울대입구(관악구청)', '낙성대(강감찬)', '사당', '방배',
    '서초', '교대(법원.검찰청)', '강남', '역삼', '선릉', '삼성(무역센터)'
]

df_route = df_pivot.reindex(station_sequence).reset_index()

df_route['Net_Inflow'] = df_route['승차'] - df_route['하차']
df_route['Cumulative_Load'] = df_route['Net_Inflow'].cumsum()

fig, ax = plt.subplots(figsize=(14, 7))

ax.plot(df_route['station'], df_route['Cumulative_Load'], marker='o', color='#e74c3c', linewidth=3, label='일평균 열차 내 누적 인원 수')
ax.fill_between(df_route['station'], df_route['Cumulative_Load'], color='#e74c3c', alpha=0.15)

ax.set_title('출근 시간대(07~09시) 2호선 주요 경로별 열차 내 누적 혼잡도 분석', fontsize=16, weight='bold', pad=20)
ax.set_xlabel('지하철 운행 경로 (외곽 주거지 → 도심 업무지)', fontsize=12, labelpad=10)

ax.set_ylabel('일평균 열차 내 누적 인원 수 (상대지수)', fontsize=12, labelpad=10)

ax.set_xticklabels(df_route['station'], rotation=30, ha='right')
ax.grid(True, linestyle='--', alpha=0.5)

ax.axvspan('신림', '교대(법원.검찰청)', color='yellow', alpha=0.2)
ax.text(3.5, df_route['Cumulative_Load'].max() * 0.85, '승객 누적 포화 구간\n(지속적 승차로 좌석 확보 불가)', color='crimson', weight='bold', fontsize=11)
ax.text(11, df_route['Cumulative_Load'].max() * 0.4, '중심 업무지 지구 진입\n(대규모 하차 및 좌석 확보 가능)', color='navy', weight='bold', fontsize=11)

lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='upper right', fontsize=11)

plt.tight_layout()

plt.savefig('fig05_commute_congestion_flow.png', dpi=300)
plt.show()
plt.close()

print("fig05_commute_congestion_flow.png 저장 완료!")
````
